## Let's Start

```python
company = soup.find_all('div',class_='company-content-wrapper')#it will get all the content on that container
```

to get all the 30 compny we use a for loop to get the all container data

company[0].find_all('p',class_='infoEntity')[0].text.strip()
#info entity have same four things present in it

```python
name=[]
rating=[]
review=[]
ctype=[]
hq=[]
old=[]
employees=[]
desc=[]

for i in company:
    name.append(i.find_all('h2')[0].text.strip())
    rating.append(i.find_all('p',class_='rating')[0].text.strip())#we can use only find to get the result beacuse this container only single
    review.append(i.find('a',class_='review-count').text.strip())
    ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())#multiple value so we use find all
    hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    desc.append(i.find('p',class_='description').text.strip())
    
#now creating data frame
d={'name':name,'rating':rating,'review':review,'ctype':ctype,'hq':hq,'old':old,'employees':employees,'desc':desc}
pd.DataFrame(d)
    
    
```

## Creating final data frame of all the pages in that website

```python
import numpy as np
```

```python
final_DataFrame=pd.DataFrame()

for j in range(1,150):
    url='https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}'.format(j)
    headers={'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'}

    webpage = requests.get(url,headers=headers).text
    
    soup=BeautifulSoup(webpage,'lxml')
    
    company = soup.find_all('div',class_='company-content-wrapper')
    
    name=[]
    rating=[]
    review=[]
    ctype=[]
    hq=[]
    old=[]
    employees=[]
    desc=[]

    for i in company:
        name.append(i.find_all('h2')[0].text.strip())
        rating.append(i.find_all('p',class_='rating')[0].text.strip())#we can use only find to get the result beacuse this container only single
        review.append(i.find('a',class_='review-count').text.strip())
        ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())#multiple value so we use find all
        
        info_entities = i.find_all('p', class_='infoEntity')
        
        if len(info_entities) > 1:
            hq.append(info_entities[1].text.strip())
        else:
            hq.append('N/A')
            
        if len(info_entities)>2:
            old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
        else:
            old.append('N/A')

        #info_entities = i.find_all('p',class_='infoEntity')
        if len(info_entities) > 3:
            employees.append(info_entities[3].text.strip())
        else:
            employees.append('N/A')

        desc_elem = i.find('p',class_='description')
        if desc_elem is not None:
            desc.append(desc_elem.text.strip())
        else:
            desc.append('N/A')


        

    #now creating data frame
    d={'name':name,'rating':rating,'review':review,'ctype':ctype,'hq':hq,'old':old,'employees':employees,'desc':desc}
    df=pd.DataFrame(d)
    
    final_DataFrame=final_DataFrame.append(df,ignore_index=True)
    
    
```
